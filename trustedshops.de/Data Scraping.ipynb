```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import os
import re
```

```python
# result = requests.get("https://www.trustedshops.de/bewertung/info_XCCB82C7D588CF17E7A10E49930AC8323.html?stars=1")
# src = result.content

#https://bugs.chromium.org/p/chromedriver/issues/detail?id=4121#c18

options = webdriver.ChromeOptions()
#options.headless = True
options.add_argument('--incognito')
driver = webdriver.Chrome(options=options)

dic = {"page":[], "review":[], "rating":[], "written":[], "edited":[], "purchased":[]}

stars = 1
for page in range(1,2):
    driver.get("https://www.trustedshops.de/bewertung/info_XCCB82C7D588CF17E7A10E49930AC8323.html?stars="+str(stars)+"&page="+str(page))
    src = driver.page_source
    soup = BeautifulSoup(src, 'lxml')
    values = soup.find_all("div", {"class" : "sc-5cbc7755-0 sc-94f04969-0 glqDZO fGfcsh"})

    for value in values:
        dic["page"].append(page)
        dic["rating"].append(stars)
        review = value.find_all("div", {"class" : "sc-94f04969-1 dafdRI"})
        if len(review) != 0:
            dic["review"].append(review[0].text)
        else:
            dic["review"].append("")

        date = value.find_all("div", {"class" : "sc-ff34bd69-2 jQdVaJ"})
        for val in date:
            date_p_e = val.find_all("span", {"class" : "sc-ff34bd69-3 jcaTTN"})
            
            if len(date_p_e) == 1:
                dic["edited"].append("")
                dic["purchased"].append("")
                if "verfasst am" in date_p_e[0].text:
                    dic["written"].append(date_p_e[0].text)
                else:
                    dic["written"].append("")

            elif len(date_p_e) == 2:
                dic["purchased"].append("")
                if "verfasst am" in date_p_e[0].text:
                    dic["written"].append(date_p_e[0].text)
                else:
                    dic["written"].append("")
                if "bearbeitet am" in date_p_e[1].text:
                    dic["edited"].append(date_p_e[1].text)
                else:
                    dic["edited"].append("")

            elif len(date_p_e) == 3:
                if "verfasst am" in date_p_e[0].text:
                    dic["written"].append(date_p_e[0].text)
                else:
                    dic["written"].append("")
                if "bearbeitet am" in date_p_e[1].text:
                    dic["edited"].append(date_p_e[1].text)
                if "gekauft am" in date_p_e[2].text:
                    dic["purchased"].append(date_p_e[2].text)
                else:
                    dic["purchased"].append("")

            else:
                dic["written"].append("else")
                dic["edited"].append("else")
                dic["purchased"].append("else")
```
