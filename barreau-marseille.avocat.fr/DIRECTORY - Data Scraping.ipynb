```python
dic = {"name":[], "address":[], "tel1":[], "fax":[], "email":[], "date":[],
                    "address2":[], "fax2":[], "email2":[],
                    "address3":[], "fax3":[], "email3":[],
                    "address4":[], "fax4":[],
                    "data":[]}

for page in range(1,100):
    result = requests.get("https://www.barreau-marseille.avocat.fr/fr/annuaire/page-"+str(page)+".html")
    src = result.content
    soup = BeautifulSoup(src, 'lxml')
    values = soup.find_all("div", {"class" : "b-annuaire__invisiblebloc"})
    for value in values:
        name = value.find_all("h2")[0].text
        dic["name"].append(name)

        #####
        data_list = []
        for val in value.find_all("p"):
            data_list.append(val.get_text(strip=True))

        while("" in data_list) :
            data_list.remove("")

        for i, item in enumerate(data_list, start=0):
            pattern=re.compile("([\n])|([\ ]{3,})")
            data_list[i] = re.sub(pattern,"",str(item))
        #####

        dic["address"].append(data_list[0])
        data_list.remove(data_list[0])

        for data in data_list:
            if "Tél" in data:
                dic["tel1"].append(data)
                data_list.remove(data)
                break
            if "Tél" not in data:
                dic["tel1"].append("")
                break
        
        for data in data_list:
            if "Fax" in data:
                dic["fax"].append(data)
                data_list.remove(data)
                break
            if "Fax" not in data:
                dic["fax"].append("")
                break

        for data in data_list:
            if "Email" in data:
                dic["email"].append(data)
                data_list.remove(data)
                break
            if "Email" not in data:
                dic["email"].append("")
                break

        for data in data_list:
            if "Début d'activité" in data:
                dic["date"].append(data)
                data_list.remove(data)
                break
            if "Début d'activité" not in data:
                dic["date"].append("")
                break

        if len(data_list) > 0 :
            dic["address2"].append(data_list[0])
            data_list.remove(data_list[0])
        else :
            dic["address2"].append("")

        if len(data_list) > 0 :
            for data in data_list:
                if "Fax" in data:
                    dic["fax2"].append(data)
                    data_list.remove(data)
                    break
                if "Fax" not in data:
                    dic["fax2"].append("")
                    break
        else :
            dic["fax2"].append("")

        if len(data_list) > 0 :
            for data in data_list:
                if "Email" in data:
                    dic["email2"].append(data)
                    data_list.remove(data)
                    break
                if "Email" not in data:
                    dic["email2"].append("")
                    break
        else :
            dic["email2"].append("")

        if len(data_list) > 0 :
            dic["address3"].append(data_list[0])
            data_list.remove(data_list[0])
        else :
            dic["address3"].append("")

        if len(data_list) > 0 :
            for data in data_list:
                if "Fax" in data:
                    dic["fax3"].append(data)
                    data_list.remove(data)
                    break
                if "Fax" not in data:
                    dic["fax3"].append("")
                    break
        else :
            dic["fax3"].append("")

        if len(data_list) > 0 :
            for data in data_list:
                if "Email" in data:
                    dic["email3"].append(data)
                    data_list.remove(data)
                    break
                if "Email" not in data:
                    dic["email3"].append("")
                    break
        else :
            dic["email3"].append("")

        if len(data_list) > 0 :
            dic["address4"].append(data_list[0])
            data_list.remove(data_list[0])
        else :
            dic["address4"].append("")

        if len(data_list) > 0 :
            for data in data_list:
                if "Fax" in data:
                    dic["fax4"].append(data)
                    data_list.remove(data)
                    break
                if "Fax" not in data:
                    dic["fax4"].append("")
                    break
        else :
            dic["fax4"].append("")

        dic["data"].append(data_list)
        
        print(page,  name)
```
